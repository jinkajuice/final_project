```python
# Make BERT results into DataFrame
results_dicts = [benchmark_bert_finer, benchmark_bert_music, bert_base_finer, bert_large_finer, distilbert_finer, bert_base_music, distilbert_music]
df_bert = pd.DataFrame(results_dicts)
df_bert['k-samples'] = "n/a"

# Add GPT results to DataFrame
gpt_dicts = [gpt_4omini_test10, gpt_4o_32_finer, gpt_4o_64_finer, gpt_4o_128_finer, gpt_4omini_32_music, gpt_4o32_music, gpt_4o64_music]
df_gpt = pd.DataFrame(gpt_dicts)

# Concatenate DataFrames
df = pd.concat([df_bert, df_gpt], ignore_index=True)
```

```python
# Sort DF by model_type, dataset and macro_f1 score
df_sorted = df.sort_values(by=['model_type','dataset','k-samples','eval_macro_f1'], ascending=[True, True, True, False])
df_sorted
```

```python
# Drop columns to show only macro scores
drop_columns = ['eval_loss', 'eval_micro_precision', 'eval_micro_recall', 'eval_micro_f1', 'eval_accuracy', 'eval_runtime', 'eval_samples_per_second', 'eval_steps_per_second', 'epoch']
df_macro_f1 = df_sorted.drop(drop_columns, axis=1)

df_macro_f1
```

```python
# Get subset of bert results
df_bert = df_macro_f1[df_macro_f1["model_type"]=="bert"] 
df_bert = df_bert.drop(['k-samples'], axis=1)
df_bert
```

```python
# Plot macro F1 scores for BERT models on finer dataset
sns.set_theme()
# Plot graph
fig, ax = plt.subplots(figsize=(6, 4))
sns.barplot(x='model', y='eval_macro_f1', data=df_bert, hue='dataset', legend=True)

# Limit y-axis to highlight differences
ax.set_ylim(0.75, 1.0) 

# Label graph
ax.set_title('Macro F1 Scores of BERT Models')
ax.set_xlabel('Model')
ax.set_ylabel('Macro F1 score')
ax.legend(title='Dataset', loc='lower left')

# Rotate x-labels
plt.xticks(rotation=45)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.4f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom')

plt.show()
```
